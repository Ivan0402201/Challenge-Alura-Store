### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

#1. Análisis de facturación



```python
def ingreso_total():
 ingresosTienda1 = tienda['Precio'].sum()
 ingresosTienda2 = tienda2['Precio'].sum()
 ingresosTienda3 = tienda3['Precio'].sum()
 ingresosTienda4 = tienda4['Precio'].sum()


 print(f'El ingreso de la tienda 1 es de: {ingresosTienda1}')
 print(f'El ingreso de la tienda 2 es de: {ingresosTienda2}')
 print(f'El ingreso de la tienda 3 es de: {ingresosTienda3}')
 print(f'El ingreso de la tienda 4 es de: {ingresosTienda4}')
 return [ingresosTienda1, ingresosTienda2, ingresosTienda3, ingresosTienda4]
ingreso_total()
```

# 2. Ventas por categoría

```python
def ventas_por_categoria():
  ventastienda1 = tienda.groupby('Categoría del Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False)
  ventastienda2 = tienda2.groupby('Categoría del Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False)
  ventastienda3 = tienda3.groupby('Categoría del Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False)
  ventastienda4 = tienda4.groupby('Categoría del Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False)


  print(f'Las ventas por categoría de la tienda 1 son: {ventastienda1}')
  print('===================================================================')
  print(f'Las ventas por categoría de la tienda 2 son: {ventastienda2}')
  print('===================================================================')
  print(f'Las ventas por categoría de la tienda 3 son: {ventastienda3}')
  print('===================================================================')
  print(f'Las ventas por categoría de la tienda 4 son: {ventastienda4}')




ventas_por_categoria()
```

# 3. Calificación promedio de la tienda


```python
def calificacion_promedio():
  calificacionTienda1 = tienda['Calificación'].mean()
  calificacionTienda2 = tienda2['Calificación'].sum() / len(tienda2['Calificación'])
  calificacionTienda3 = tienda3['Calificación'].mean()
  calificacionTienda4 = tienda4['Calificación'].mean()

  print(f'La calificación promedio de la tienda 1 es de: {calificacionTienda1:.2f}')
  print(f'La calificación promedio de la tienda 2 es de: {calificacionTienda2:.2f}')
  print(f'La calificación promedio de la tienda 3 es de: {calificacionTienda3:.2f}')
  print(f'La calificación promedio de la tienda 4 es de: {calificacionTienda4:.2f}')
  return [calificacionTienda1, calificacionTienda2, calificacionTienda3, calificacionTienda4]

calificacion_promedio()
```

# 4. Productos más y menos vendidos

```python
def productos_mas_vendidos():
  productosTienda1 = tienda['Producto'].value_counts().head()
  productosTienda2 = tienda2['Producto'].value_counts().head()
  productosTienda3 = tienda3['Producto'].value_counts().head()
  productosTienda4 = tienda4['Producto'].value_counts().head()

  print(f'Los productos más vendidos de la tienda 1 son: {productosTienda1}')
  print('===================================================================')
  print(f'Los productos más vendidos de la tienda 2 son: {productosTienda2}')
  print('===================================================================')
  print(f'Los productos más vendidos de la tienda 3 son: {productosTienda3}')
  print('===================================================================')
  print(f'Los productos más vendidos de la tienda 4 son: {productosTienda4}')

productos_mas_vendidos()
```

```python
def productos_menos_vendidos():
  productosTienda1 = tienda['Producto'].value_counts().tail()
  productosTienda2 = tienda2['Producto'].value_counts().tail()
  productosTienda3 = tienda3['Producto'].value_counts().tail()
  productosTienda4 = tienda4['Producto'].value_counts().tail()

  print(f'Los productos menos vendidos de la tienda 1 son: {productosTienda1}')
  print('===================================================================')
  print(f'Los productos menos vendidos de la tienda 2 son: {productosTienda2}')
  print('===================================================================')
  print(f'Los productos menos vendidos de la tienda 3 son: {productosTienda3}')
  print('===================================================================')
  print(f'Los productos menos vendidos de la tienda 4 son: {productosTienda4}')

productos_menos_vendidos()
```

# 5. Envío promedio por tienda

```python
def envio_promedio():
  envioTienda1 = tienda['Costo de envío'].mean()
  envioTienda2 = tienda2['Costo de envío'].sum() / len(tienda2['Costo de envío'])
  envioTienda3 = tienda3['Costo de envío'].mean()
  envioTienda4 = tienda4['Costo de envío'].mean()

  print(f'El envío promedio de la tienda 1 es de: {envioTienda1:.2f}')
  print(f'El envío promedio de la tienda 2 es de: {envioTienda2:.2f}')
  print(f'El envío promedio de la tienda 3 es de: {envioTienda3:.2f}')
  print(f'El envío promedio de la tienda 4 es de: {envioTienda4:.2f}')
  return [envioTienda1, envioTienda2, envioTienda3, envioTienda4]

envio_promedio()
```

```python
def grafico_barras():

  import matplotlib.pyplot as plt

  tiendasx = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
  ventasy = ingreso_total()
  ventasy_millones = [v / 1_000_000 for v in ventasy]


  plt.bar(tiendasx, ventasy_millones)


  plt.title('Ventas totales por tienda (en millones)')
  plt.xlabel('Tiendas')
  plt.ylabel('Ventas (millones de pesos)')

  # Mostrar valores sobre cada barra
  for i, valor in enumerate(ventasy_millones):
    plt.text(i, valor + 10, f"${valor:,.0f}M", ha='center', va='bottom', fontsize=10)

  plt.show()

grafico_barras()
```

```python
def grafico_dispersion():


  import matplotlib.pyplot as plt

  tiendasx = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
  promedioy = calificacion_promedio()



  plt.scatter(tiendasx, promedioy)


  plt.title('Calificacion por tienda')
  plt.xlabel('Tiendas')
  plt.ylabel('Calificacion')

  plt.show()

grafico_dispersion()
```

```python
def grafico_linea():

  import matplotlib.pyplot as plt

  tiendasx = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
  ventasy = envio_promedio()



  plt.plot(tiendasx, ventasy)


  plt.title('Envio Promedio por tienda')
  plt.xlabel('Tiendas')
  plt.ylabel('Costo Promedio x Envio')

  plt.show()

grafico_linea()
```

# **Introducción**

El objetivo de este análisis es proporcionar al Sr. Juan una recomendación informada sobre cuál tienda debe elegir para vender sus productos, considerando múltiples factores clave: ingresos totales, ventas por categoría, calificaciones promedio de clientes, productos más y menos vendidos, y costos de envío promedio. Se analizaron cuatro tiendas con datos de ventas y satisfacción del cliente para identificar fortalezas y debilidades de cada una.


## **1. Ingresos Totales**
* La tienda 1 registra el mayor ingreso total con 1,150,880,400.0 unidades
monetarias.

* Le siguen la tienda 2 (1,116,343,500.0), la tienda 3 (1,098,019,600.0) y la tienda 4 (1,038,375,700.0).

Esto indica que la tienda 1 lidera en volumen de ventas monetarias, pero las diferencias no son muy amplias entre tiendas.



## **2. Ventas por Categoría**
En todas las tiendas, la categoría Muebles es la más vendida, seguida de cerca por Electrónicos y Electrodomésticos o Juguetes, dependiendo de la tienda.

Las categorías con menor venta incluyen Instrumentos musicales, Libros y Artículos para el hogar.

La tienda 1 muestra la mayor cantidad de cuotas (indicador indirecto de ventas) en "Muebles" (1446) y en "Electrónicos" (1341), posicionándola como fuerte en esas categorías.



## **3. Calificaciones Promedio de Clientes**
La tienda 3 tiene la mejor calificación promedio con 4.05.

La tienda 2 le sigue con 4.04, la tienda 4 con 4.00, y la tienda 1 registra la calificación más baja con 3.98.

Esto sugiere que la satisfacción del cliente es más alta en las tiendas 2 y 3, aspecto fundamental para fidelización y reputación.

## **4. Productos Más y Menos Vendidos**
La tienda 1 destaca en ventas de productos como Microondas, TV LED UHD 4K y Armarios, con 60 unidades vendidas en los primeros.

La tienda 2 lidera con libros como "Iniciando en programación" y productos musicales como Guitarra acústica.

La tienda 3 muestra preferencia en mobiliario como Kit de bancas y Mesa de comedor.

La tienda 4 tiene fuertes ventas en camas y artículos como Cama box y Cubertería.

En cuanto a productos menos vendidos, cada tienda tiene una variedad que incluye desde productos tecnológicos hasta muebles, lo cual puede indicar oportunidades para mejorar el catálogo.

## **5. Costos de Envío Promedio**
La tienda 4 tiene el costo de envío promedio más bajo (23,459.46), seguida por la tienda 3 (24,805.68), tienda 2 (25,216.24) y tienda 1 con el costo más alto (26,018.61).

Un costo de envío menor puede traducirse en mejores márgenes y competitividad en precio.

Conclusión y Recomendación
Considerando todos los factores analizados, la recomendación para el Sr. Juan es vender en la Tienda 3.

## **Justificación:**

Aunque la tienda 1 tiene el mayor ingreso total, presenta la calificación promedio más baja y el costo de envío más alto, lo que puede afectar la satisfacción y rentabilidad a largo plazo.

*La tienda 3 combina un ingreso competitivo, la mejor calificación promedio de clientes (4.05), y uno de los costos de envío más bajos, aspectos clave para una operación sostenible y con buena percepción de los compradores.*

**Además, su oferta de productos más vendidos está alineada con muebles y artículos para el hogar, que son categorías fuertes y estables.**

**La tienda 3 también muestra un buen balance en ventas por categoría, con un público diverso y satisfecho.**

Si bien la tienda 2 también tiene buena calificación y ventas, su ingreso total es menor que el de la 3, y el costo de envío ligeramente más alto. La tienda 4, pese a su bajo costo de envío, tiene el menor ingreso y calificación más baja que la 3 y 2.

```python
import pandas as pd

# Agrega una columna para saber de qué tienda vienen los datos
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Unimos todos los DataFrames
df_total = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

df_total[['Tienda', 'lat', 'lon', 'Precio', 'Calificación']].head()

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_total, x='lon', y='lat', hue='Tienda', alpha=0.6)

plt.title('Distribución geográfica de ventas por tienda')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend()
plt.grid(True)
plt.show()


import folium
from folium.plugins import HeatMap

# Crear un mapa centrado en la media de las coordenadas
centro_mapa = [df_total['lat'].mean(), df_total['lon'].mean()]
mapa = folium.Map(location=centro_mapa, zoom_start=6)

# Agregar capa de calor (heatmap)
datos_heatmap = df_total[['lat', 'lon']].dropna().values.tolist()
HeatMap(datos_heatmap, radius=10).add_to(mapa)

mapa




```
